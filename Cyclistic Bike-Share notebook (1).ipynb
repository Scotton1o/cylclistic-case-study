## Average ride length.

In SQL calculate ride_length then calculate average ride_length grouped by member_casual and grouped by month
```sql
-- Step A: Add the new column
ALTER TABLE `temptables-465123.case_study.cyclistic_2025`
ADD COLUMN ride_length INT64;

-- Step B: Populate with the difference in seconds
UPDATE `temptables-465123.case_study.cyclistic_2025`
SET ride_length = TIMESTAMP_DIFF(ended_at, started_at, SECOND)
WHERE started_at IS NOT NULL AND ended_at IS NOT NULL;

SELECT 
    member_casual, 
    EXTRACT(MONTH FROM started_at) AS ride_month,

WHERE ride_length IS NOT NULL
GROUP BY 
    member_casual, 
    ride_month

## Rides by Month and Day.

Total number of rides taken broken down by months and days of the week grouped by member_casual. Looking for seasonal trends.
In SQL I added a column named “day_of_week” listing the correct day based on values in the “started_at” column. To calculate start day.

````r
#R Code
library(ggplot2)
library(dplyr)
library(readr)

# File path updated for a generic Jupyter environment
df_rides <- read_csv("ride_count.csv",
                     show_col_types = FALSE)

df_summer <- df_rides %>%
  filter(month %in% c("June", "July", "August")) %>%
  mutate(
    # Ensure months follow calendar order
    month = factor(month, levels = c("June", "July", "August")),
    # Ensure days of week follow calendar order
    day_of_week = factor(day_of_week, levels = c("Monday", "Tuesday", "Wednesday",
                                                  "Thursday", "Friday", "Saturday", "Sunday"))
  )

ggplot(df_summer, aes(x = day_of_week, y = ride_count, fill = member_casual)) +
  geom_col(position = "dodge") +
  facet_wrap(~ month) +
  labs(title = "Summer Ride Counts: June - August 2025",
       subtitle = "Comparative analysis of member vs. casual usage by day of week") +
  # --- Add this line to tilt x-axis labels by 45 degrees ---
  theme(axis.text.x = element_text(angle = 45, hjust = 1))
````

## Top ten start stations

Provide data on top ten start_station_name for casual riders. 
This sql code gives Top 10 start_station_name for casual riders on day_of_week Saturday and Sunday during months of June, July, Aug. Without including Null values 
```sql
SELECT 
    start_station_name, 
    COUNT(ride_id) AS ride_count
FROM temptables-465123.case_study.cyclistic_2025
WHERE 
    -- Filter for casual riders
    member_casual = 'casual'
    -- Filter for specific months (6=June, 7=July, 8=August)
    AND EXTRACT(MONTH FROM started_at) IN (6, 7, 8)
    -- Filter for weekend days
    AND day_of_week IN ('Saturday', 'Sunday')
    -- Exclude null values from the grouping column and ID column
    AND start_station_name IS NOT NULL
    AND ride_id IS NOT NULL
GROUP BY 
    start_station_name
ORDER BY 
    ride_count DESC
LIMIT 10;

# Share
## What story does this data tell us?


## Rider Base and Conversion Potential
Annual members currently outnumber casual riders by 28%. This variance highlights a significant opportunity to develop targeted campaigns aimed at converting the casual rider segment into annual members.

## Rideable Type Preferences
Both rider demographics demonstrate a preference for electric bikes. Marketing strategies should prioritize ensuring high availability and prominent promotion of the electric fleet.

## Usage Patterns and Timing

Casual riders generally take longer average trips and exhibit significantly higher maximum ride lengths specifically during March and June.
The sole periods where casual riders surpass members in ride count are on Saturdays and Sundays in the summer months (June, July, August). The optimal time to market to this demographic is explicitly during summer weekends.
## Location-Based Targeting

Popular start station locations differ markedly between groups.
Casual riders favor stations along the eastern lakeshore.
Annual members prefer stations within the inner city and southern areas near Hyde Park.
The most effective location-based marketing campaigns should be deployed along the eastern lakeshore to capture casual rider attention.
